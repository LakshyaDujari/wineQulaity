```python
scaler = StandardScaler()
temp = scaler.fit_transform(red_wine_final.iloc[:,:-1])
```

```python
x_train,x_test,y_train,y_test = train_test_split(temp,red_wine_final.iloc[:,-1])
```

```python
from sklearn.metrics import mean_squared_error,r2_score

lreg = LinearRegression()
lreg.fit(x_train,y_train)
print(lreg.score(x_test,y_test))
print("important feature in linear modal:\n",lreg.coef_)
# Make predictions on the data
predictions = lreg.predict(x_test)

# Calculate the MSE
mse = mean_squared_error(y_test, predictions)

print("Mean Squared Error:", mse)
# Calculate R-squared using scikit-learn
r_squared = r2_score(y_test, predictions)

print("R-squared (R) using scikit-learn:", r_squared)
```

```python
logModal = LogisticRegression()
logModal.fit(x_train,y_train)
print("Method 1:",logModal.score(x_test,y_test))
y_pred = logModal.predict(x_test)
print("Method 2:",accuracy_score(y_test,y_pred))
```
